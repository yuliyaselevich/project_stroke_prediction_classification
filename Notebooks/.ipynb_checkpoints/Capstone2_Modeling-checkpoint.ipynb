# Modeling

```python
# import necessary libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import pickle
from matplotlib.ticker import PercentFormatter
import seaborn as sns
from scipy.stats import loguniform
from sklearn.metrics import (roc_auc_score, precision_score,
                             recall_score, f1_score,fbeta_score)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV, cross_validate, GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report,roc_curve, confusion_matrix,brier_score_loss
import xgboost as xgb
```

```python
import warnings
warnings.filterwarnings('ignore', category=UserWarning, append=True)
warnings.filterwarnings('ignore', category=RuntimeWarning, append=True)
warnings.filterwarnings('ignore', category=FutureWarning, append=True)
warnings.filterwarnings('ignore',category=pd.io.pytables.PerformanceWarning)
```

```python
# Load train and test sets 
X_train = pd.read_csv('../Data/X_train.csv',index_col=0)
X_test = pd.read_csv('../Data/X_test.csv',index_col=0)
y_train = pd.read_csv('../Data/y_train.csv',index_col=0)
y_test = pd.read_csv('../Data/y_test.csv',index_col=0)
```

```python
# Reset the index of the training and testing dataframes and labels to start from 0
X_train.reset_index(drop=True,inplace=True)
X_test.reset_index(drop=True,inplace=True)
y_train.reset_index(drop=True,inplace=True)
y_test.reset_index(drop=True,inplace=True)
```

```python
# Flatten the label arrays using the ravel() function
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
# Print the shape of the feature and label arrays for training and testing sets
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## StratifiedKFold cross-validator ensures that each fold of the dataset has the same proportion of observations with a given label, therefore it appropriate to use it for this imbalanced dataset.

```python
# Define a StratifiedKFold object for cross-validation
cv = StratifiedKFold(n_splits= 5, shuffle=True, random_state=42)
```

## Logistic regression

```python
# Define a dictionary of hyperparameters to be used in Logistic Regression model
lr_parameters = {'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'], 
              'C':[100, 10, 1.0, 0.1, 0.01],
             'penalty':['none', 'l1', 'l2', 'elasticnet']}
```

```python
# Create an instance of the Logistic Regression model
lr_model = LogisticRegression()
# Create an instance of the GridSearchCV class
# The GridSearchCV function will perform a cross-validation grid search 
# over the hyperparameters specified in the dictionary 'lr_parameters'
lr_clf = GridSearchCV(lr_model, param_grid = lr_parameters,cv=cv,scoring='roc_auc')
```

```python
# Fit the GridSearchCV object to the training data
#lr_clf.fit(X_train, y_train)
```

```python
#print('The best score is', lr_clf.best_score_,'with following best parameters',lr_clf.best_params_)
```

```python
# Instantiate and fit a LogisticRegression model with specific hyperparameters on the training data
lr_model = LogisticRegression(C=10, penalty='elasticnet', 
                              solver='saga',l1_ratio = 0.1,random_state=50).fit(X_train,y_train)
# Predict probabilities of the target class (1) on the test set using the fitted model
lr_probs = lr_model.predict_proba(X_test)[:, 1]
# Predict target variable (binary) on the test set using the fitted model
y_pred = lr_model.predict(X_test)
```

```python
print(f'Logistic Regression\n')
# Calculate and print the ROC AUC score for the logistic regression model on the test set
roc = roc_auc_score(y_test, lr_probs)
print(f'ROC AUC: {round(roc, 4)}')
print(classification_report(y_test, y_pred,zero_division=0))
```

```python
# Create a confusion matrix with predicted and actual values
cf_matrix = confusion_matrix(y_test,y_pred)
ax =sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')
ax.set_title('Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])
plt.show()
```

```python
# Create a confusion matrix with predicted and actual values
cf_matrix = confusion_matrix(y_test,xgb_preds)
ax =sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')
ax.set_title('Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

## Plotting ROC curves for all three models.

```python
# Set the figure size
fig = plt.gcf()
fig.set_size_inches(18, 10, forward=True)
# Define the models to be compared
models = [(lr_model, 'Logistic Regression'),(rf_model, 'Random Forest Classifier'),(xgb_model, 'XGBoost Classifier')]
# Iterate over each model and plot its ROC curve
for model, name in models:
    # Make predictions on the test data
    probs = model.predict_proba(X_test)[:, 1]
    # Calculate the AUC score
    auc = roc_auc_score(y_test, probs)
    # Calculate the ROC curve
    fpr, tpr, _ = roc_curve(y_test, probs)
    # Calculate the Brier score
    br = brier_score_loss(y_test, probs, pos_label=1)
    label = f'{name} AUC: {round(auc, 3)} Brier Score: {round(br, 3)}'
    plt.plot(fpr, tpr, marker='.', label=label)
# Plot the no-skill line
plt.plot([0, 1], [0, 1], linestyle='--', label='No Skill')
plt.xlabel('False Positive Rate', fontsize=15)
plt.ylabel('True Positive Rate', fontsize=15)
plt.title('ROC curves comparison', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)
plt.show()
```

## Even though XGBoost Classifier has a slightly higher ROC-AUC score than Logistic Regression, I will use Logistic Regression as my final model as it has better brier score.

## Calculating presicion-recall based on different beta thresholds.

```python
# Define a function to calculate precision, recall, and F1 score for a given threshold
def thr(threshold):
    y_pred = (lr_model.predict_proba(X_test)[:,1] >= threshold)
    p = precision_score(y_test,y_pred)
    r = recall_score(y_test,y_pred)
    f = fbeta_score(y_test, y_pred,beta=1)
    return p,r,f
```

```python
# Define an empty dictionary to store the scores for each threshold
thr_dict = {}
# Iterate over each unique threshold in the logistic regression predictions
for threshold in sorted(list(set(lr_model.predict_proba(X_test)[:,1]))):
    p,r,f = thr(threshold)
    thr_dict[threshold] = p,r,f
```

```python
# Transform resulting dictionary to a DataFrame
res = pd.DataFrame.from_dict(thr_dict).T.rename(columns={"index": "Threshold", 0: "Precision",
                                                                              1: "Recall",2:'F1_score'})
res
```

```python
### Adapted from https://machinelearningmastery.com/threshold-moving-for-imbalanced-classification/
def to_labels(pos_probs, threshold):
    return (pos_probs >= threshold).astype('int')
yhat = lr_model.predict_proba(X_test)
# Keep probabilities for the positive outcome only
probs = yhat[:, 1]
# Define thresholds
thresholds = np.arange(0, 1, 0.001)
# Evaluate each threshold
scores = [fbeta_score(y_test, to_labels(probs, t),beta=1) for t in thresholds]
# Get best threshold
ix = np.argmax(scores)
print('Threshold=%.3f, F-Score=%.5f' % (thresholds[ix], scores[ix]))
```

## Considering that Type 2 error is more important than Type 1 error in case of prediction of health conditions, we are going to favor higher recall over precision. Therefore optimal threshold 

```python
# Create a plot of precision and recall scores against threshold values
ax = res.plot(figsize=(15,9))
ax.set_xlabel('Threshold')
ax.set_ylabel('Value')
ax.set_title('Precision-Recall vs. Threshold')
ax.axvline(0.5, color ='black', alpha = 0.3)
ax.text(.485,.5,'Default Threshold',rotation=90, alpha=0.8, fontsize=16)
ax.axvline(0.117, color ='black', alpha = 0.3)
ax.text(.103,.5,'Optimal Threshold',rotation=90, alpha=0.8, fontsize=16)
plt.xlabel('False Positive Rate',fontsize=15)
plt.ylabel('True Positive Rate',fontsize=15)
plt.title('Precision-Recall vs. Threshold',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)
plt.show()
```
